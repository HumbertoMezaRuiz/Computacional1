# **Problema 1**

image.png

```python
#De la ecuación de movimiento definimos    x'  = v
#Despejamos la segunda deriada como        x'' = -2 * C * w * v - w**2 * x + (1/m) * F * np.sin(w * t)

# Definimos la función F(t,y), el lado derecho de la ecuacion diferencial
# La solución y es un arreglo 1-D, y=(theta, omega), dy/dt es un arreglo

def oscilador_amortiguado(y, t, C, m, w, F):
        x, v = y
        dydt = [v, -2 * C * w * v - w**2 * x + (1/m) * F * np.sin(w * t)]
        return dydt

# Definimos el tiempo de integración de 0 a 50 segundos (5001 puntos, h=0.01)
t = np.linspace(0, 7, 1001)

## Definimos las condiciones iniciales x=-1, e v = 0 (el oscilador se "suelta") y las constantes
y0 = [-1.0, 0.0]
F = 0.0
m = 1.0
k = 9
w = np.sqrt(k/m)
```

```python
## PRIMER CASO C=0.0
C = 0.0
y = odeint(oscilador_amortiguado, y0, t, args=(C,m,w,F))
## SEGUNDO CASO C=0.1
C=0.1
y2 = odeint(oscilador_amortiguado, y0, t, args=(C,m,w,F))
## TERCER CASO C=0.2
C=0.2
y3 = odeint(oscilador_amortiguado, y0, t, args=(C,m,w,F))
## CUARTO CASO C=0.4
C=0.4
y4 = odeint(oscilador_amortiguado, y0, t, args=(C,m,w,F))
## QUINTO CASO C=0.7
C=0.7
y5 = odeint(oscilador_amortiguado, y0, t, args=(C,m,w,F))
## SEXTO CASO C=1.0
C=1
y6 = odeint(oscilador_amortiguado, y0, t, args=(C,m,w,F))
## SÉPTIMO CASO C=2.0
C=2
y7 = odeint(oscilador_amortiguado, y0, t, args=(C,m,w,F))

plt.figure(figsize=(10,5))
plt.plot(t, y[:, 0] , 'black', label='C=0.0')
plt.plot(t, y2[:, 0], 'g', label='C=0.1')
plt.plot(t, y3[:, 0], 'b', label='C=0.2')
plt.plot(t, y4[:, 0], 'r', label='C=0.4')
plt.plot(t, y5[:, 0], 'orange', label='C=0.7')
plt.plot(t, y6[:, 0], 'purple', label='C=1.0')
plt.plot(t, y7[:, 0], 'gray', label='C=2.0')
plt.legend(loc='best')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.grid()
plt.show()
```
