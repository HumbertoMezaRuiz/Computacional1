```
import numpy as np
import matplotlib.pyplot as plt

from scipy.integrate import odeint
```

```
## Definimos la primera derivada como x' = v
## Despejamos la segunda derivada como x'' = u*v + u*v*x^2-x


from scipy.integrate import odeint

# Definimos la función F(t,y), el lado derecho de la ecuacion diferencial
# La solución y es un arreglo 1-D, y=(theta, omega), dy/dt es un arreglo

def pendulo(y, t, u):
        x, v = y
        dydt = [v, u*v - u*x**2*v-x]
        return dydt

# Definimos el tiempo de integración de 0 a 50 segundos (5001 puntos, h=0.01)
t = np.linspace(0, 50, 5001)

## Definimos las condiciones iniciales theta=0.3, y omega = 0 (el péndulo se "suelta")
y0 = [0.3, 0.0]
```

```
## PRIMER CASO U=0
u = 0.0

y = odeint(pendulo, y0, t, args=(u,))

plt.figure(figsize=(10,5))
plt.plot(t, y[:, 0], 'b', label='x')
plt.plot(t, y[:, 1], 'g', label='v')
plt.legend(loc='best')
plt.title('Solución con mu=0')
plt.xlabel('t')
plt.grid()
plt.show()

## PRIMER CASO U=1
u = 1.0

y = odeint(pendulo, y0, t, args=(u,))

plt.figure(figsize=(10,5))
plt.plot(t, y[:, 0], 'b', label='x(t)')
plt.plot(t, y[:, 1], 'g', label='v(t)')
plt.legend(loc='best')
plt.title('Solución con mu=1')
plt.xlabel('t')
plt.grid()
plt.show()

## PRIMER CASO U=2
u = 2.0

y = odeint(pendulo, y0, t, args=(u,))

plt.figure(figsize=(10,5))
plt.plot(t, y[:, 0], 'b', label='x(t)')
plt.plot(t, y[:, 1], 'g', label='v(t)')
plt.legend(loc='best')
plt.title('Solución con mu=2')
plt.xlabel('t')
plt.grid()
plt.show()

## PRIMER CASO U=3
u = 3.0

y = odeint(pendulo, y0, t, args=(u,))

plt.figure(figsize=(10,5))
plt.plot(t, y[:, 0], 'b', label='x(t)')
plt.plot(t, y[:, 1], 'g', label='v(t)')
plt.legend(loc='best')
plt.title('Solución con mu=3')
plt.xlabel('t')
plt.grid()
plt.show()

## PRIMER CASO U=4
u = 4.0

y = odeint(pendulo, y0, t, args=(u,))

plt.figure(figsize=(10,5))
plt.plot(t, y[:, 0], 'b', label='x(t)')
plt.plot(t, y[:, 1], 'g', label='v(t)')
plt.legend(loc='best')
plt.title('Solución con mu=4')
plt.xlabel('t')
plt.grid()
plt.show()
```
